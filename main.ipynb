### Understanding the data
I asked chatpgt what the dataset could represent:

The first few columns are basic stuff as: date, open, high, low, close, adjclose, volume, ticker. These are apparantly standard columns in stock related databases, ticker is the name of the stock dat is used.
<br>
Then we have some RSI (relative strength index columns): RSIadjclose15, RSIvolume15, RSIadjclose25, RSIvolume25, RSIadjclose50, RSIvolume50
<br>
<br>RSI (Relative Strength Index) — simple explanation
<br>The RSI is a number between 0 and 100 that tells you whether a stock’s price has been going up or down too quickly.
<br>If RSI is high (above 70) → the stock has gone up a lot recently → it might be overbought (too expensive).
<br>If RSI is low (below 30) → the stock has gone down a lot recently → it might be oversold (too cheap).
<br>So RSI is basically a momentum score that shows if prices are climbing or falling faster than usual.
<br>
<br>Let’s unpack one:
<br>Column name	Meaning
<br>RSIadjclose15	RSI of the stock’s adjusted closing price, calculated over the last 15 days
<br>RSIvolume15	RSI of the trading volume (how many shares were traded) over the last 15 days
<br>RSIadjclose25	Same idea, but over 25 days
<br>RSIadjclose50	Same idea, but over 50 days
<br>

From this I can determine that in the whole database the column name consists of a type, for example RSI, a type of volume and an amount of days for this statistic.<br>
For me the most important words to understand in this database are: RSI, MACD, Lag, Sentiment, Moving averages.<br>
- RSI: Relative Strength Index, this is a value between 0 and 100, this value determines how much a stock has gone up in a certain time period. A higher value means more difference. 
- MACD: This value represents how much the change of the change of a stock is happening. It is the differential of the amount of change so to say.
- Lag: This is what the value used to be a timestamp ago, most likely a day.
- Sentiment: This is how investors feel about the stock, for example greedy, fearful or another sentiment.
- Moving averages: The moving average is an average of a few values in a certain time that changes when a new timestamp is added.


#### What could be important for my dataset

Close price – most critical, reflects the actual market consensus at the end of the period.<br>
Moving averages (short-term, e.g., 5- or 10-day) – capture short-term trends and momentum.<br>
MACD – shows the difference between short-term and long-term trends; good for spotting momentum shifts.<br>
RSI / Stochastic Oscillator – indicate overbought or oversold conditions, useful for reversals.<br>
Volume – confirms strength of price moves; high volume often signals continuation or reversal.<br>
Hammer / candlestick patterns – less frequent, but can indicate reversal points when they appear.<br>
Lags (previous prices or returns) – help models capture patterns over time; their importance grows with time-series models.<br>
Sentiment – adds context for sudden moves driven by news or social mood; useful, but often noisier than technical indicators.<br>
So, for most predictive models, price-based indicators, moving averages, and momentum indicators carry the most weight, while candlestick patterns and sentiment act as supportive signals.<br>
